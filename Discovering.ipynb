```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as snb
```

```python
forest_df = pd.read_csv('./Forest_Data.csv')
forest_df.head()
```

```python
# First Would like to have a plot of total world's forest percentage varying with time.
world_percentage_data = forest_df.loc[0, ['Forest Percentage(1990)','Forest Percentage(2000)','Forest Percentage(2010)','Forest Percentage(2015)','Forest Percentage(2020)',]]
```

```python
world_percentage_data
```

```python
type(world_percentage_data)
```

```python
world_percentage_data.index
```

```python
year_list = [1990, 2000, 2010, 2015, 2020]

plt.scatter(np.array(year_list), np.array(world_percentage_data), marker='o', color='r')



plt.plot(np.array(year_list), np.array(world_percentage_data))
plt.xlabel('Year')
plt.ylabel('Total Percentage of forest in the world')
plt.title('Forest percentage Evolution')
plt.grid(True)
plt.show()
```
